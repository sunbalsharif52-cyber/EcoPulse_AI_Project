```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

print("🌐 Yahoo Finance API se Live Crude Oil data fetch kiya ja raha hai...")

# 1. Aaj ki date aur pichle 3 saal ki date automatic calculate karna
end_date = datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.today() - timedelta(days=3*365)).strftime('%Y-%m-%d')

# 2. Brent Crude Oil ka official global ticker (BZ=F) use karna
ticker = "BZ=F" 
oil_data = yf.download(ticker, start=start_date, end=end_date)

# 3. Data ko check aur clean karna
if not oil_data.empty:
    print(f"✅ Data successfully downloaded! Total records: {len(oil_data)}")
    # Sirf Close (Closing Price) aur Volume columns ko track karna
    oil_df = oil_data[['Close', 'Volume']].copy()
    
    print("\n--- DATA PREVIEW (Top 5 Rows) ---")
    print(oil_df.head())
else:
    print("❌ API se data fetch nahi ho saka. Internet connection check karein.")
```
